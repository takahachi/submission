```python
#grouplens

import pandas as pd
import numpy as np

def pearson(x, y): #ピアソンの相関係数(重み)
    w_1, w_2 = [], []
    tmp = x * y
    for i in range(len(tmp)):
        if tmp[i] > 0 and tmp[i] <= 25 :
            w_1.append(x[i])
            w_2.append(y[i])
    
    return np.corrcoef(w_1, w_2)[0,1]

def average(bar):
    total,fig = 0, 0
    for i in bar:
        if i!=-9999:
            total+=i
            fig+=1
    return total/fig

def evaluate(a,i,trainig,weighit):
    w, tmp =0, 0
    for u in range(training.shape[0]):
        if a!=u and training[u,i]!=-9999 :
            w+=abs(weight[a][u])
            tmp+= (training[u,i]-average(training[u]))*weight[a][u]
    
    return average(training[a])+tmp/w


training_df = pd.read_csv("kinki_training.csv",  encoding="shift-jis", header=None, index_col=0)
test_df = pd.read_csv("kinki_test.csv",  encoding="shift-jis", header=None,index_col=0)
training_df = training_df.drop(index=["回答者番号"])
test_df = test_df.drop(index=["回答者番号"])
training = training_df.to_numpy()

#重み
weight = [[0 for i in range(training.shape[0])]for j in range(training.shape[0])]
#予測値
predict= [[0 for i in range(training.shape[1])]for j in range(training.shape[0])]

#training.shape(70, 21) ユーザ数×アイテム数
training = training.astype('int16')

for a in range(training.shape[0]):
    for u in range(training.shape[0]):
        if a==u:
            continue    
        weight[a][u] = pearson(training[a], training[u])
     
for a in range(training.shape[0]):
    for i in range(training.shape[1]):
        predict[a][i] = evaluate(a,i,training,weight)
    
predict_df = pd.DataFrame(predict)
predict_df
weight_df = pd.DataFrame(weight)
#weight_df
```

```python
#パーソナライズされていない手法

def evaluate_noper_onlyave(a,i,training):
    tmp, n = 0, 0
    for u in range(training.shape[0]):
        if training[u,i]!=-9999 :
            tmp+=training[u,i]
            n+=1
    return tmp/n

def evaluate_noper_ave(a,i,training):
    tmp, n = 0, 0
    for u in range(training.shape[0]):
        if training[u,i]!=-9999 :
            tmp+=training[u,i]-average(training[u])
            n+=1

    return average(training[a])+tmp/n


predict_noper_onave = [[0 for i in range(training.shape[1])]for j in range(training.shape[0])]
predict_noper_ave = [[0 for i in range(training.shape[1])]for j in range(training.shape[0])]

for a in range(training.shape[0]):
    for i in range(training.shape[1]):
        predict_noper_onave[a][i] =  evaluate_noper_onlyave(a,i,training)
        predict_noper_ave[a][i] = evaluate_noper_ave(a,i,training)
        
predict_noper_onave_df = pd.DataFrame(predict_noper_onave)
predict_noper_ave_df = pd.DataFrame(predict_noper_ave)
predict_noper_onave_df
```

```python
#平均絶対誤差(MAE)
import math

def MAE(test,predict):
    tmp,n=0,0
    for j in range(test.shape[0]):
        if not math.isnan(predict[j][test[j,0]-1]):
            tmp +=abs(predict[j][test[j,0]-1]-test[j,1])
            n+=1
    return tmp/n

test = test_df.to_numpy()
test = test.astype("int8")
mae_1=MAE(test,predict)#grouplens
mae_2=MAE(test,predict_noper_onave)#パーソナライズされていない手法（予測値として平均値）
mae_3=MAE(test,predict_noper_ave)#パーソナライズされていない手法（予測値としてユーザごとの平均からの偏差）

print("パーソナライズされた手法")
print("grouplensでの予測の平均絶対誤差：",mae_1)
print("\nパーソナライズされていない手法")
print("予測値としてすべてのユーザの評価値の平均値を用いたものの平均絶対誤差：",mae_2)
print("予測値としてユーザごとの平均値からの偏差を用いたものの平均絶対誤差：",mae_3)
```
